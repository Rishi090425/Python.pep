```python
with open("int.txt", "w") as file:
    file.write("My name is Rishi Kumar\n")
    file.write("I'm pursuing my Btech degree at LPU\n")
file.close()
with open("int.txt", "a") as file:
    file.write("I am learning Python.\n")
with open("int.txt", "r") as file:
    print(file.read())

with open("int.txt", "r") as file:
    lines = file.read()
    print("Number of lines:", len(lines))
```

```python
with open("new.txt", "w") as file:
    file.write("Today is friday")
    
w = input("Enter word: ")
with open("new.txt", "r") as file:
    txt = file.read()

if w in txt:
    print("Word found")
else:
    print("Word not found")
```

```python
with open("new.txt", "w") as file:
    file.write("Today is friday")
with open("new.txt", "r") as file:
    txt = file.read()
    words = txt.split()

print("Number of words:", len(words))
```

```python

with open("int.txt", "r") as file:
    data = file.read()

with open("upper.txt", "w") as file:
    file.write(data.upper())

with open("upper.txt", "r") as file:
    print(file.read())

print("File converted to uppercase successfully.")
    
```

```python
with open("students.txt", "w") as file:
    for i in range(5):
        name = input("student name: ")
        file.write(name + "\n")

print("Done")
```

```python
with open("students.txt", "r") as file:
    lines = file.readlines()

for i in range(0, len(lines), 2):
    print(lines[i], end="")
```

```python
class Student:
    college="LPU"

    def __init__(self,name,age):
        self.name=name
        self.age=age

    def display(self):
        print(self.name,self.age,self.college)

s1=Student("Rahul",20)
s2=Student("Amit",21)
s1.display()
s2.display()
```
