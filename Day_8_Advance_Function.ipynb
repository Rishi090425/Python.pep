```python
from functools import reduce
numbers = [2,4,6,3,5]
largest = reduce(lambda x, y: x if x > y else y, numbers)
print("Largest number is:", largest)
```

```python
numbers = [3,5,2,6,7]
cube = map(lambda x: x**3, numbers)
print(list(cube))
```

```python
celsius = [22,65,4,34,45]
fahrenheit = map(lambda c: (c * 9/5) + 32, celsius)
print(list(fahrenheit))
```

```python
numbers = [34, 56, 765, 54, 5, 3]
even_numbers = map(lambda x: "even" if x % 2 == 0 else "odd", numbers)
print(list(even_numbers))
```

```python
numbers = [1,-4,2,-5,4,-2]
positive_negative = map(lambda x: "positive" if x>=0 else "negative", numbers)
print(list(positive_negative))
```

```python
num = int(input("Enter a number: "))
table= map(lambda n: f"{num} x {n} = {num*n}", range(1, 11))
print(list(table))
```
