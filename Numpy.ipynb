```python
import numpy as np

marks = np.array([[87,88,89,90,91],[95,98,87,89,92],[76,65,67,72,73],[99,98,89,92,93],[54,56,67,62,60]])

avg = np.mean(marks)
print(avg)
```

```python
highest = np.max(marks)
lowest = np.min(marks)

print(highest)
print(lowest)
```

```python
std = np.std(marks)
print(std)
```

```python
print(np.where(marks> avg)[0])
```

```python
percentage = (np.sum(marks)/500)*100 
print(percentage)
```
