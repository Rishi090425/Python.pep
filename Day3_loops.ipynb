```python
# palindrome number

num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
# reversing a number
num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed number is:", rev)
```

```python
# skip multiple of 3 till 20
i = 1

while i <= 20:
    if i % 3 == 0:
        i = i + 1
        continue

    print(i)
    i = i + 1
```
