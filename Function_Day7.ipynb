```python
Quantity = int(input("Enter quantity: "))
Amount = int(input("Enter Amount: "))

def final_bill(Quantity, Amount):
    if Amount >= 10000 and Quantity >= 1:
        return Amount + Amount * 12 / 100
    elif Amount >= 5000 and Amount < 10000 and Quantity >= 1:
        return Amount + Amount * 8 / 100
    else:
        return "No Discount"

print(final_bill(Quantity, Amount))
```
