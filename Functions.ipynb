```python

def add(a, b):
    return a + b
    
def subtract(a, b):
    return a - b


def multiply(a, b):
    return a * b


def divide(a, b):
    return a / b

print(add(5, 6))
print(subtract(5, 9))
print(multiply(7, 3))
print(divide(6, 7))



```

```python
def largest(a, b):
    if a > b:
        return a, "is the largest"
    elif b > a:
        return b, "is the largest"
    else:
        return "Both numbers are equal"

print(largest(5, 7))
print(largest(2, 1))
print(largest(4, 4))
```

```python
def even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_odd(10))
print(even_odd(7))
print(even_odd(0))
```

```python
def largest(a, b, c, d, e):
    if a >= b and a >= c and a >= d and a >= e:
        return a
    elif b >= a and b >= c and b >= d and b >= e:
        return b
    elif c >= a and c >= b and c >= d and c >= e:
        return c
    elif d >= a and d >= b and d >= c and d >= e:
        return d
    else:
        return e

print(largest(10, 25, 7, 18, 3))
print(largest(5, 5, 5, 5, 5))
print(largest(12, 9, 30, 4, 11))
```

```python
def add(*num):
    return sum(num)

print(add(23, 22, 5))
print(add(26, 54, 37, 33))
```

```python
def student(**info):
    return info

print(student(name="Ritik", age=20, branch="CSE"))
```
